```python
import pandas as pd
```

```python
books = pd.read_csv("clean.csv")
```

```python
books1 = books.copy()
```

```python
books = books1.drop(["property_categories", "property_isbn 10", "property_isbn 13"], axis = 1)
```

```python
books.loc[books.property_pages == 8827, "property_pages"] = 827
books.loc[books.property_pages.isna(), "property_pages"] = int(books.property_pages.median())
books.loc[books.property_year.isna(), "property_year"] = int(books.property_year.median())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Boxplots pour les variables cat√©goriques

```python
plt.figure(figsize = (15,5))
plt.subplot(1,3,1)
sns.boxplot(x = 'property_edition', y = 'popularity', data = books)
plt.subplot(1,3,2)
sns.boxplot(x = 'property_language', y = 'popularity', data = books)
plt.subplot(1,3,3)
sns.boxplot(x = 'file_type', y = 'popularity', data = books)
```

## Violinplots pour les variables num√©riques

```python
plt.figure(figsize = (15,5))
plt.subplot(1,3,1)
plt.violinplot(books.property_pages)
plt.subplot(1,3,2)
plt.violinplot(books.file_size)
plt.subplot(1,3,3)
plt.violinplot(books.property_year)
plt.show()
```

## Diagrammes

```python
languages = list(books.property_language.value_counts().index)
n = len(books)
data_languages = []
for language in languages:
    data_languages.append((100/n)*len(books[books.property_language == language]))
labels_languages = ['Others']+languages[1:]
```

```python
categories = list(books.columns[10:-1])
data_categories = []
for category in categories:
    data_categories.append(len(books[books[category] == 1]))
```

```python
editions = list(books.property_edition.value_counts().index)
data_editions = []
for edition in editions:
    data_editions.append((100/n)*len(books[books.property_edition == edition]))
labels_editions = editions
```

```python
file_types = list(books.file_type.value_counts().index)
data_file_types = []
for file_type in file_types:
    data_file_types.append((100/n)*len(books[books.file_type == file_type]))
```

```python
plt.figure(figsize = (12,12))
plt.subplot(2,2,1)
plt.pie(data_languages, labels=labels_languages, autopct='%1.1f%%', startangle=90)
plt.subplot(2,2,2)
plt.pie(data_categories, labels=categories, autopct='%1.1f%%', startangle=90)
plt.subplot(2,2,3)
plt.pie(data_editions, labels=editions, autopct='%1.1f%%', startangle=90)
plt.subplot(2,2,4)
plt.pie(data_file_types, labels=file_types, autopct='%1.1f%%', startangle=90)
plt.show()
```

## R√©partiton de la moyenne des nombres des pages et la taille des fichiers par rapport au types

```python
pages = [int(books.loc[books.file_type == ft, "property_pages"].mean()) for ft in books.file_type.value_counts().index]
sizes = [int(books.loc[books.file_type == ft, "file_size"].mean()) for ft in books.file_type.value_counts().index]
```

```python
import numpy as np
plt.figure(figsize = (10,6))
X = np.arange(5)
plt.bar(X + 0.00, pages, color = 'b', label="x", width = 0.25)
plt.bar(X + 0.25, sizes, color = 'r', label="y", width = 0.25)
```

## Kde taille des fichiers / popularit√©
